```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
#make one for pH and lead
```

```python
df = pd.DataFrame({'Contaminants':['Iron', 'Mercury', 'Copper', 'Aluminum', 'Sulfate', 'Bromine'], 'Parts Per Million':[10 ,20, 30 , 40 ,50 ,60]})
ax = df.plot.bar(x='Contaminants', y='Parts Per Million', rot=0)
```

```python
leadData = pd.DataFrame({"Lead": [0, 1,3, 5,15,2]})
leadData['Lead'].plot(kind="bar", title = 'Lead')


plt.title("Amount of Lead found in water")
plt.xlabel("Sample")
plt.ylabel("Lead parts ber billion")
```

```python
pHData = pd.DataFrame({"pH": [6.0, 6.5, 6.7, 7.5, 8.0]})
pHData['pH'].plot(kind="bar", title = 'pH')


plt.title("pH Level")
plt.xlabel("Sample")
plt.ylabel("Level of pH")
```
